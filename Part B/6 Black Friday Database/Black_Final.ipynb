```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("blackfri.csv")
df.head(15)
```

```python
df.describe()
```

```python
df.drop(['User_ID','Product_ID','Stay_In_Current_City_Years'],axis = 1,inplace = True)
df.head()
```

```python
df['City_Category'].fillna('A',inplace = True)
df
```

```python
df.rename(columns= {
    "Product_Category_1" : "Baseball Caps",
    "Product_Category_2" : "Baseball Caps",
    "Product_Category_3" : "Baseball Caps"
})
```

```python
df["City_Category"] = df["City_Category"].map({
    'A' : "Metro",
    'B' : "Small Town",
    'C' : "Villages"
})
```

```python
df.head()
```

```python
df['Gender'] = df['Gender'].map({
    'F' : "Female",
    'M' : "Male"
})

df['Marital_Status'] = df['Marital_Status'].map({
    0 : "Un-married",
    1 : "Married"
})
```

```python
df.head()
df.info()
```

```python
ndf = pd.DataFrame(data = df,columns=['Gender','Product_Category_1','Product_Category_2'])
ndf.dropna(subset = ["Product_Category_2"],inplace = True)
ndf.groupby("Gender").count()
```

```python
ax = sns.countplot(data = ndf,x = "Gender")
```

```python
ax = sns.countplot(data = df,x = "Gender",hue = "City_Category")
ax.set(title = "Total Number of Male & Female persons belonging to each city category")
plt.show()
```
