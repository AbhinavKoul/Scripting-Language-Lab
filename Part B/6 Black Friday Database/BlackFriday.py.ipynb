<b>c) Remove unnecessary features (E.g. drop unwanted columns) from the dataset such as ‘User_ID’, ‘Product_ID ‘ ‘Stay_In_Current_City_Years</b>

```python
df.drop(['User_ID','Product_ID','Stay_In_Current_City_Years'], axis=1, inplace=True)
df.head(5)
```

<b>d) Manipulate data by replacing empty column values in ‘City_Category’ with a default value for the city.</b>

```python
print("Filling empty values")
df['City_Category'] = df['City_Category'].fillna("A")
df.head(5)
```

```python
print("Mapping values/attributes in City_Category to types")
df['City_Category'] = df['City_Category'].map({'A':'Metro cities','B':'Small Towns','C':'Villages'})
df.head(5)
```

<b>f) Rename the attribute ‘Product_Category_1’ to have ‘Baseball Caps’,",
    "‘Product_Category_2’ to have ‘Wine Tumblers’ and ‘Product_Category_3’ to 
    "have ‘Pet Raincoats’</b>

```python
print("Renaming the column names")
df.rename(columns={'Product_Category_1':'Baseball_Caps',
                   'Product_Category_2':'Wine_Tumblers',
                   'Product_Category_3':'Pet_Raincoats'},inplace=True)
df.head(5)
```

<b>g) Convert the attribute ‘Marital_Status’ to have ‘1:Married’ and ‘0:Un-Married’</b>

```python
print("Mapping values/attributes in Marital Status to types")
df['Marital_Status'] = df['Marital_Status'].map({1:'Married',
                                                 0:'Un-Married'})
df.head(5)
```

<b>h) Perform the following visualizations on the loaded dataset:",  
i)Tally of the Number of Male & Female who bought‘Product_Category_3(Pet_Raincoats)’</b>

```python
import matplotlib.pyplot as plt
import seaborn as sns
print("<-------Data Visualisation------->")
print(pd.crosstab(df.Gender,df.Baseball_Caps))
print()
print(pd.crosstab(df.Gender,df.Pet_Raincoats))

ax = sns.countplot(data=df,x='Gender',hue='Pet_Raincoats',palette='Set2')
ax.set(title='Male and Female who bought Pet_Raincoats',xlabel='Gender',ylabel='Count')
plt.show()
```

<b>ii)  Total Number of Male & Female persons belonging to each city category</b>

```python
ax = sns.countplot(data=df,x='Gender',hue='City_Category',palette='Set1')
ax.set(title='Male and Female belonging to each city',xlabel='Gender',ylabel='Count')
plt.show()
```
